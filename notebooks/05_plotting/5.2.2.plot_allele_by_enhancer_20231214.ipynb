```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_data(dataframes, row_index, title):
    sns.set_palette("colorblind")
    plt.rcParams['figure.dpi'] = 150

    data_for_plot = []

    # Extracting REF and ALT values for each dataframe
    for name, df in dataframes.items():
        row = df.loc[row_index]
        ref_cols = [col for col in df.columns if col.endswith('_REF')]
        alt_cols = [col for col in df.columns if col.endswith('_ALT')]

        for ref_col, alt_col in zip(ref_cols, alt_cols):
            data_for_plot.append({'DataFrame': name, 'Type': 'REF', 'Value': row[ref_col]})
            data_for_plot.append({'DataFrame': name, 'Type': 'ALT', 'Value': row[alt_col]})

    # Create DataFrame for plotting
    plot_df = pd.DataFrame(data_for_plot)

    # First Plot: Boxplot and Stripplot
    plt.figure(figsize=(4, 6))
    sns.set(style="whitegrid", font_scale=1.2)
    
    # Create box plots and add individual data points
    ax1 = sns.boxplot(x='DataFrame', y='Value', hue='Type', data=plot_df, palette="colorblind")
    sns.stripplot(x='DataFrame', y='Value', hue='Type', data=plot_df, jitter=True, color='black', alpha=0.5, size=3, dodge=True)

    # Adjust grid settings and labels for first plot
    adjust_grid_and_labels(ax1)

    plt.title(title)
    plt.ylabel('Mad Score of Transcription Rate')
    plt.xlabel('')
    plt.show()

    # Second Plot: Stripplot with connecting lines
    plt.figure(figsize=(4, 6))
    sns.set(style="whitegrid", font_scale=1.2)

    # Define color mapping
    palette = {'REF': 'blue', 'ALT': 'orange'}

    # Create stripplot
    ax2 = sns.stripplot(x='DataFrame', y='Value', hue='Type', data=plot_df, jitter=True, dodge=True, palette=palette, alpha=0.5, size=3)

    # Draw connecting lines
    draw_connecting_lines(ax2, plot_df)

    # Adjust grid settings and labels for second plot
    adjust_grid_and_labels(ax2)

    plt.title(title)
    plt.ylabel('Mad Score of Transcription Rate')
    plt.xlabel('')
    plt.show()

def adjust_grid_and_labels(ax):
    grid_style = {'linestyle': '--', 'linewidth': 0.5, 'color': 'gray', 'alpha': 0.3}
    ax.xaxis.grid(True, **grid_style)
    ax.yaxis.grid(True, **grid_style)
    ax.set_xticklabels(ax.get_xticklabels(), fontsize=12)
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles[:2], labels[:2], title='Type', loc='upper right')

def draw_connecting_lines(ax, plot_df):
    dataframe_categories = plot_df['DataFrame'].unique()
    for dataframe in dataframe_categories:
        ref_values = plot_df[(plot_df['DataFrame'] == dataframe) & (plot_df['Type'] == 'REF')]['Value'].values
        alt_values = plot_df[(plot_df['DataFrame'] == dataframe) & (plot_df['Type'] == 'ALT')]['Value'].values
        ref_pos = ax.get_xticks()[list(dataframe_categories).index(dataframe)] - 0.2
        alt_pos = ax.get_xticks()[list(dataframe_categories).index(dataframe)] + 0.2
        for ref, alt in zip(ref_values, alt_values):
            plt.plot([ref_pos, alt_pos], [ref, alt], linestyle='--', color='gray')
```
