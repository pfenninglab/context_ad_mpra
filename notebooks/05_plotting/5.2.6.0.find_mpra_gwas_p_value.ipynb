```python
import pandas as pd

# Load the table containing "RSID" column
rsid_table = pd.read_csv('indexing/SNP_haploregannotatr_ATAC_tss_SNPFunction_spliceai_bayesianML_20240919.csv',index_col=0)
unique_rsids = set(rsid_table['RSID'].unique())  # Extract unique RSIDs and store in a set

# Initialize an empty list to store matching rows
matching_rows = []

# Define a function to process each chunk of the TSV file
def filter_chunk(chunk):
    # Filter rows in the chunk where the variant_id matches one of the RSIDs
    return chunk[chunk['variant_id'].isin(unique_rsids)]

# Load the large TSV file in chunks
chunk_size = 10**6  # Adjust chunk size if needed based on memory
tsv_file_path = '/media/zihengc/T7/mpra/resources/Bellenguez/GCST90027158_buildGRCh38.tsv'

# Iterate over chunks of the TSV file
for chunk in pd.read_csv(tsv_file_path, sep='\t', chunksize=chunk_size):
    matching_rows.append(filter_chunk(chunk))

# Concatenate all matching rows into a final DataFrame
result_df = pd.concat(matching_rows)

# Save the result if needed
result_df.to_csv('indexing/20241021_SNPs_GWAS_match.csv', index=False)
```
