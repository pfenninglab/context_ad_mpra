```python
import pandas as pd
import os

# Set your directory path
base_directory_path =  '/media/zihengc/T7/mpra3_lib_analysis/enhancer_activities'

# Function to process each file
def process_file(file_path, folder_name):
    try:
        df = pd.read_csv(file_path,index_col=0)
        if 'pval.mad' in df.columns:
            df_sorted = df[df['pval.mad']<=0.05]
            df_sorted = df.sort_values(by='pval.mad')
            top_10_indices = df_sorted.head(10).index.tolist()
            return top_10_indices
        else:
            print(f"'pval.mad' column not found in {file_path}")
            return [None] * 10
    except Exception as e:
        print(f"Error processing file {file_path}: {e}")
        return [None] * 10

# Function to process each directory
def process_directory(directory_path, folder_name):
    top_indices_dict = {}
    for filename in os.listdir(directory_path):
        if filename.endswith('.csv'):
            file_path = os.path.join(directory_path, filename)
            indices = process_file(file_path, folder_name)
            column_name = f"{folder_name}_{filename.replace('.csv', '').replace('20240105_MPRA3_MAD_merged', '')}"
            top_indices_dict[column_name] = indices
    return top_indices_dict

# Main
all_top_indices = {}

# Check each subdirectory in the base directory
for dirname in os.listdir(base_directory_path):
    dir_path = os.path.join(base_directory_path, dirname)
    if os.path.isdir(dir_path) and dirname.endswith("_byAllele"):
        all_top_indices.update(process_directory(dir_path, dirname))

# Convert the dictionary to a DataFrame
final_df = pd.DataFrame(all_top_indices)

# Set the DataFrame's index to 1 through 10
final_df.index = range(1, 11)

# Save the DataFrame to a CSV file
output_path = os.path.join(base_directory_path, 'all_top_indices_summary.csv')
final_df.to_csv(output_path)

print(f"All top indices summary saved to {output_path}")
```
