# Transformation to have ALT and REF in the same column

```python
import pandas as pd
def convert_reshaped(input,output):
    index=pd.read_csv("indexing/ALT_REF_LookUpTable_filtered_amended_20231117.csv",index_col=0)
    df = pd.read_csv(input,index_col=0)
    ref = df.loc[:,df.columns.str.contains("REF")]
    ref = pd.merge(ref,index,left_index=True,right_index=True)
    ref = ref.set_index("1")
    ref = ref.drop_duplicates().drop('PanTissueControl.SV40_224bp_1xEnh_Promoter')
    ref.rename(columns=lambda x: x.replace('REF', 'ALT'), inplace=True)
    alt = df.loc[:,df.columns.str.contains("ALT")]
    combined = pd.concat([alt,ref])
    combined = combined[combined.index.str.contains("ref") | combined.index.str.contains("alt")|combined.index.str.contains("motifdisrupt")]
    combined.to_csv(output)
```

```python
input = "read_counts_R1R2/THP1MacrophagePseudobarcodes_RNA_matched_barcodes_reshaped.csv"
output = 'read_counts_R1R2/THP1MacrophagePseudobarcodes_RNA_matched_barcodes_reshaped_altref.csv'
convert_reshaped(input,output)
input = "read_counts_R1R2/THP1MacrophagePseudobarcodes_DNA_matched_barcodes_reshaped.csv"
output = 'read_counts_R1R2/THP1MacrophagePseudobarcodes_DNA_matched_barcodes_reshaped_altref.csv'
convert_reshaped(input,output)
```
