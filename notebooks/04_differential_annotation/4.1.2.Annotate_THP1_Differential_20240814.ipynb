```python
#overlap Pfenning peaks
import pandas as pd
import os
import gzip

# Assuming df_final is already loaded and has the correct columns
df_peak = pd.read_csv('indexing/SNP_haploreg_annotated_20231204.csv', index_col=0)

# Function to check if a position is within a peak
def is_within_peak(chrom, pos, peak_df):
    # Ensure the chromosome is a string because it can be 'X', 'Y', or a number
    chrom = str(chrom)
    # Convert position to integer to perform numeric comparisons
    pos = int(pos)
    #if chrom == 11 and pos == 
    # Filter peak_df for rows where the variant is in the chromosome
    # Ensure peak_df['chrom'] is string type to match with 'chrom'
    chrom_peaks = peak_df[peak_df['chrom'].astype(str) == chrom]
    
    # Check if 'pos' is within any of the peak ranges
    # Start and end positions are converted to integers for comparison
    within_peak = chrom_peaks[(chrom_peaks['start'].astype(int) <= pos) & (chrom_peaks['end'].astype(int) >= pos)]
    
    # If 'within_peak' is not empty, return 1, else return 0
    return 1 if not within_peak.empty else 0

# Directory containing the peak and bed files
directory = "../atac/all_open_chromatin_peaks/THP1_4hrs_differential_fdr005"

# Loop through the files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".narrowPeak.gz") or filename.endswith(".bed"):
        # Construct file path
        file_path = os.path.join(directory, filename)
        
        # Read the file into a DataFrame
        if filename.endswith(".narrowPeak.gz"):
            with gzip.open(file_path, 'rt') as f:
                peaks_df = pd.read_csv(f, sep='\t', header=None, usecols=[0, 1, 2])
        else:  # .bed file
            peaks_df = pd.read_csv(file_path, sep='\t', header=None, usecols=[0, 1, 2])
        
        # Rename columns for consistency
        peaks_df.columns = ['chrom', 'start', 'end']
        
        # Remove 'chr' prefix if present
        peaks_df['chrom'] = peaks_df['chrom'].str.replace('chr', '')

        # Column name derived from the file name
        column_name = filename.split('.')[0]
        
        # Annotate variants in df_peak with peak information
        df_peak[column_name] = df_peak.apply(
            lambda row: is_within_peak(row['chr'], row['pos_hg38'], peaks_df), axis=1
        )

# Save the annotated DataFrame to a CSV file
df_peak.to_csv('indexing/variants_peak_annotation_THP1FourHrsDifferential_20240117.csv')
```

```python
import pandas as pd
def get_significant_interaction(interaction_file):
    df = pd.read_csv(interaction_file,index_col=0)
    df = df[df["fdr"]<=0.05]
    df["Comparison"] = interaction_file[21:-16]
    return df

df=pd.DataFrame()
for i in ['20240129_comparative_HMC3_IFNBvsNaive_interaction.csv',
          '20240129_comparative_HMC3_IFNGvsNaive_interaction.csv',
          '20240129_comparative_HMC3_LPSIFNGvsIFNG_interaction.csv',
          '20240129_comparative_HMC3_LPSIFNGvsNaive_interaction.csv',
          '20240129_comparative_THP1_IFNBvsNaive_interaction.csv',
          '20240129_comparative_THP1_IFNGvsNaive_interaction.csv',
          '20240129_comparative_THP1_LPSIFNGvsIFNG_interaction.csv',
          '20240129_comparative_THP1_LPSIFNGvsNaive_interaction.csv']:
    df = pd.concat([df,get_significant_interaction(i)])

df_haploreg = pd.read_csv('indexing/SNP_haploreg_annotated_20231204.csv',index_col=0)
pd.merge(df_haploreg,df,left_index=True,right_index=True).sort_index().to_csv("20240205_THP1HMC3_stimulation_specific_haploreg_annotated.csv")
```
