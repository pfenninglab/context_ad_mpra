```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load data
data = pd.read_csv("enhancer_activities_differential_great_analyses/greatExportAll_THP1_df_diff_down_SNP_majorminor_lpsifng_significant.csv")
# Prepare the data
go_bp_data = data[data['# Ontology'] == 'GO Biological Process'].sort_values('HyperFdrQ').head(10)
# Create the improved bar plot
plt.figure(figsize=(10, 6))
bars = plt.barh(go_bp_data['Desc'], -np.log10(go_bp_data['HyperFdrQ']), color='mediumseagreen')
# Add labels and title
plt.xlabel('-log10(fdr)', fontsize=12, labelpad=10)
plt.title('GO Biological Processes Enriched in LPSIFNG Activation', fontsize=14, pad=20)
# Invert y-axis for descending order
plt.gca().invert_yaxis()
# Adjust layout for better spacing
plt.tight_layout()
# Save the plot with high resolution
plt.savefig('go_enrichment_lpsifng.pdf', dpi=300)  # You can increase dpi value for even higher resolution
# Show the plot
plt.show()
```
