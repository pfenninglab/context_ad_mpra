```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr, pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

df = merged_df.copy()

use_abs = True
cap_y_at_05 = False  # 你原来是 y<=0.5；不想截断就改 False

# x: this MPRA, y: Cooper et al. MPRA
x_raw = df["logFC_converted"].to_numpy(dtype=float)
y_raw = df["MPRA_corrected_logFC"].to_numpy(dtype=float)

x = np.abs(x_raw) if use_abs else x_raw
y = np.abs(y_raw) if use_abs else y_raw

mask = np.isfinite(x) & np.isfinite(y)
x, y = x[mask], y[mask]

# ---- stats ----
rho, p_rho = spearmanr(x, y)
r, p_r = pearsonr(x, y)

X = x.reshape(-1, 1)
model = LinearRegression().fit(X, y)
slope = float(model.coef_[0])
intercept = float(model.intercept_)
y_hat = model.predict(X)
r2 = r2_score(y, y_hat)

def fmt_p(p):
    return "<1e-99" if p < 1e-99 else f"{p:.2e}"

annot = (
    f"Linear: y = {slope:.3f}x + {intercept:.3f}\n"
    f"Pearson r = {r:.3f}, P = {fmt_p(p_r)}\n"
    f"$R^2$ = {r2:.3f}\n"
     f"Spearman ρ = {rho:.3f}, P = {fmt_p(p_rho)}\n"
    f"N = {len(x)}"
)

# ---- plot ----
fig, ax = plt.subplots(figsize=(7.4, 6.2))
sns.scatterplot(
    x=x, y=y, ax=ax,
    s=18, alpha=0.55, linewidth=0
)

ax.set_xlabel("absolute logFC (This MPRA)", fontsize=12)
ax.set_ylabel("absolute log2FC (Cooper et al. MPRA)", fontsize=12)
ax.set_title("HEK293T MPRA", fontsize=13)

# ---- axis limits (robust) ----
# x upper bound: robust quantile so outliers don't squash the plot
x_hi = np.quantile(x, 0.995) if len(x) > 20 else np.max(x)
x_hi = max(x_hi * 1.05, 0.5 if use_abs else 0)

ax.set_xlim(0, x_hi) if use_abs else ax.set_xlim(np.quantile(x, 0.005), np.quantile(x, 0.995))

if use_abs:
    if cap_y_at_05:
        ax.set_ylim(0, 0.5)
    else:
        y_hi = np.quantile(y, 0.995) if len(y) > 20 else np.max(y)
        ax.set_ylim(0, max(y_hi * 1.05, 0.5))
else:
    # signed view: symmetric is nicer for direction concordance
    lim = max(np.max(np.abs(x)), np.max(np.abs(y))) * 1.05
    ax.set_xlim(-lim, lim)
    ax.set_ylim(-lim, lim)

# ---- regression line ----
xmin, xmax = ax.get_xlim()
xx = np.linspace(xmin, xmax, 200)
ax.plot(xx, slope * xx + intercept, linewidth=2.2, label="Linear regression")

# ---- statistics box (TOP RIGHT) ----
ax.text(
    0.98, 0.98, annot,
    transform=ax.transAxes,
    va="top", ha="right",
    fontsize=10,
    bbox=dict(boxstyle="round,pad=0.4", facecolor="white", alpha=0.92, linewidth=0.8),
)

# ---- legend (TOP RIGHT, below stats) ----
ax.legend(frameon=False, loc="upper right", bbox_to_anchor=(0.98, 0.72))

ax.grid(True, linewidth=0.5, alpha=0.22)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr, pearsonr
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# ---- data ----
eps = 1e-300
x_raw = merged_df["fdr"].to_numpy(dtype=float)
y_raw = merged_df["MPRA_FDR"].to_numpy(dtype=float)
x = -np.log10(np.clip(x_raw, eps, 1.0))
y = -np.log10(np.clip(y_raw, eps, 1.0))

mask = np.isfinite(x) & np.isfinite(y)
x, y = x[mask], y[mask]

# ---- stats ----
rho, p_rho = spearmanr(x, y)
r, p_r = pearsonr(x, y)

X = x.reshape(-1, 1)
model = LinearRegression().fit(X, y)
slope = float(model.coef_[0])
intercept = float(model.intercept_)
y_hat = model.predict(X)
r2 = r2_score(y, y_hat)

def fmt_p(p):
    return "<1e-99" if p < 1e-99 else f"{p:.2e}"

annot = (
    f"Linear: y = {slope:.3f}x + {intercept:.3f}\n"
    f"Pearson r = {r:.3f}, P = {fmt_p(p_r)}\n"
    f"$R^2$ = {r2:.3f}\n"

    #f"Spearman ρ = {rho:.3f}, P = {fmt_p(p_rho)}\n"
    f"N = {len(x)}"
)

# ---- plot ----
fig, ax = plt.subplots(figsize=(7.2, 6.2))
sns.scatterplot(x=x, y=y, ax=ax, s=18, alpha=0.65, linewidth=0)

ax.set_xlabel("-log10(FDR) this MPRA", fontsize=12)
ax.set_ylabel("-log10(MPRA_FDR) Cooper et al. MPRA", fontsize=12)
ax.set_title("HEK293T MPRA", fontsize=13)

# symmetric limits
xmin, xmax = np.min(x), np.max(x)
ymin, ymax = np.min(y), np.max(y)
lo = min(xmin, ymin)
hi = max(xmax, ymax)
pad = 0.03 * (hi - lo) if hi > lo else 1.0
lo -= pad
hi += pad
ax.set_xlim(lo, hi)
ax.set_ylim(lo, hi)

xx = np.linspace(lo, hi, 200)
#ax.plot(xx, xx, linestyle="--", linewidth=1, label="y = x")
ax.plot(xx, slope * xx + intercept, linewidth=2, label="Linear regression")

# ---- statistics box: TOP RIGHT ----
ax.text(
    0.98, 0.98, annot,
    transform=ax.transAxes,
    va="top", ha="right",
    fontsize=10,
    bbox=dict(boxstyle="round,pad=0.4", facecolor="white", alpha=0.9, linewidth=0.8),
)

# ---- legend: ALSO TOP RIGHT, but slightly lower to avoid overlap ----
ax.legend(
    frameon=False,
    loc="upper right",
    bbox_to_anchor=(0.98, 0.62),  # adjust this if your stats box is taller/shorter
)

ax.grid(True, linewidth=0.5, alpha=0.25)
plt.tight_layout()
plt.show()
```
