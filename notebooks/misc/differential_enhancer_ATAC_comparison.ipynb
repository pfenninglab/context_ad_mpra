```python
import pandas as pd
```

```python
def find_overlap_peak(df,df_atac):

    df_haploreg = pd.read_csv('/media/zihengc/T7/mpra3_lib_analysis/indexing/enhancer_haploregannotatr_ATAC_tss_SNPFunction_20240911.csv',index_col=0)
    df = pd.merge(df,df_haploreg,left_index=True,right_index=True)
    df = df[df['fdr']<=0.05]
    mpra_df = df[['chr','pos_hg38','fdr','logFC','start','end','summit'
]]


    df_atac = df_atac.rename({"Unnamed: 0":"peak"},axis=1)
    atac_df = df_atac[df_atac['padj']<=0.05]
    # Parse ATAC peak column to chr, start, end
    atac_df[['chr', 'start_end']] = atac_df['peak'].str.split(':', expand=True)
    atac_df[['start', 'end']] = atac_df['start_end'].str.split('-', expand=True)
    atac_df['start'] = atac_df['start'].astype(int)
    atac_df['end'] = atac_df['end'].astype(int)
    atac_df.drop(columns=['start_end'], inplace=True)

    # Creating a list to store overlapping information with the updated requirements
    overlapping_info_with_index = []

    # Iterate over MPRA dataframe and check for overlaps with updated information
    for index, mpra_row in mpra_df.iterrows():
        chr_match = atac_df[atac_df['chr'] == mpra_row['chr']]
        overlap = chr_match[(chr_match['start'] <= mpra_row['pos_hg38']) & (chr_match['end'] >= mpra_row['pos_hg38'])]
        
        if not overlap.empty:
            for _, atac_row in overlap.iterrows():
                overlapping_info_with_index.append({
                    "mpra_index": index,
                    "chr": mpra_row['chr'],
                    "pos_hg38": mpra_row['pos_hg38'],
                    "mpra_logFC": mpra_row['logFC'],
                    "atac_peak": atac_row['peak'],
                    "atac_log2FoldChange": atac_row['log2FoldChange'],
                    "atac_padj": atac_row['padj'],
                    'mpra_peak':str(mpra_row['start'])+':'+str(mpra_row['end'])+':'+str(mpra_row['summit']),
                })

    # Convert overlapping information to DataFrame
    overlapping_table_with_index_df = pd.DataFrame(overlapping_info_with_index)

    # Convert overlapping information to DataFrame
    return overlapping_table_with_index_df
```

```python
df = pd.read_csv('/media/zihengc/T7/mpra3_lib_analysis/enhancer_activities_differential/20240626_comparative_THP1_LPSIFNGvsNaive_DifferentialEnhancer.csv' ,index_col=0)
df_atac = pd.read_csv('/media/zihengc/T7/atac/differential_analysis/novaseq2/data/bam/THP1_LPS4_IFNB4_IFNG4_NAIVE/LPSIFNG4_vs_Naive_IHW.csv' )
lps_naive = find_overlap_peak(df,df_atac)
lps_naive
```
